```python
from mlxtend.frequent_patterns import apriori,association_rules
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("basket_sets.csv")
```

```python
df
```

```python
freqitems = apriori(df, min_support=0.07, use_colnames=True)
```

```python
freqitems
```

```python
ar = association_rules(freqitems, metric="lift", min_threshold=1)
```
